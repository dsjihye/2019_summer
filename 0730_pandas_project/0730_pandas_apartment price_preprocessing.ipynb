#### 상권배후지- 아파트 파일

```python
# 데이터 적재 및 가격분위별 정보 추가

# 상권배후지- 아파트 파일
아파트 = pd.read_excel("서울시 우리마을가게 상권분석서비스(상권배후지-아파트).xls")
아파트2018 = 아파트[아파트['기준년월 코드']>201800]
아파트2018['아파트 평균 시가'].sort_values
percentile = np.percentile(아파트2018['아파트 평균 시가'], [0, 25, 50, 75, 100], interpolation='nearest').tolist()
아파트2018['아파트평균시가_분위']=0

###### 아파트2018['아파트 평균 시가'][0]
for i in range(len(아파트2018)):
    if 아파트2018['아파트 평균 시가'][i]<percentile[1]:
        아파트2018['아파트평균시가_분위'][i]=1
    elif 아파트2018['아파트 평균 시가'][i]<percentile[2]:
        아파트2018['아파트평균시가_분위'][i]=2
    elif 아파트2018['아파트 평균 시가'][i]<percentile[3]:
        아파트2018['아파트평균시가_분위'][i]=3
    else:
        아파트2018['아파트평균시가_분위'][i]=4
아파트2018.rename(columns={'기준년월 코드':'기준년월코드', '상권 코드':'상권코드'})
아파트2018.to_csv('아파트_2018.csv')
```

###  집객시설 아파트분위수 추가

```python
아파트2018 = pd.read_csv('data/서울시상권데이터/아파트_2018.csv',index_col=0)
```

```python
집객시설2018['아파트시가_분위'] = None
```

### 아파트 분위 추가 

```python
for idx in range(len(집객시설2018)):
    
    날짜 = 집객시설2018.loc[idx,'기준_년월_코드']
    상권코드 = 집객시설2018.loc[idx,'상권_코드']
    try:
        분위 = 아파트2018[(아파트2018['상권 코드'] == 상권코드) & (아파트2018['기준년월 코드'] == 날짜)].iloc[0]['아파트평균시가_분위']
    
    except Exception: 
        분위 = 아파트2018[아파트2018['상권 코드'] == 상권코드].iloc[0]['아파트평균시가_분위']
#    print(idx, 날짜, 상권코드, 분위)
    집객시설2018.loc[idx,'아파트시가_분위'] = 분위
    
    
```

### 아파트 가격 추가 

```python
집객시설2018['아파트 평균 시가'] = None
```

```python
from tqdm import tqdm_notebook
for idx in tqdm_notebook(range(len(집객시설2018))):
    
    날짜 = 집객시설2018.loc[idx,'기준_년월_코드']
    상권코드 = 집객시설2018.loc[idx,'상권_코드']
    try:
        분위 = 아파트2018[(아파트2018['상권 코드'] == 상권코드) & (아파트2018['기준년월 코드'] == 날짜)].iloc[0]['아파트 평균 시가']
    
    except Exception: 
        가격 = None
#    print(idx, 날짜, 상권코드, 분위)
    집객시설2018.loc[idx,'아파트 평균 시가'] = 가격
```
